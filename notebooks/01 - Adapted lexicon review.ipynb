# DALX

## Amazon Electronics

```python
corpus = load_semeval15_laptop('data/SemEval15_laptop/train.csv', 'data/SemEval15_laptop/test.csv')
vocab = read_yaml('data/SemEval15_laptop/glove_symdict.yml')
atlx = pd.read_csv('data/processed/lexicon_table_v2.csv', index_col='WORD')
atlx['ATLX'] = atlx.mean(axis=1).apply(lambda p: -1 if p < 0 else (0 if p==0 else 1))
dalx = pd.read_csv('data/output/lexicon_table_dalx_07_thres0.7_C10.csv', index_col='WORD')
diff_df = atlx.join(dalx[['DALX']])
diff_df['INVOCAB'] = diff_df.index.isin(vocab)
```

```python
# Words in S15 vocab
diff_df.INVOCAB.value_counts() # 839
```

```python
# Words that changed polarity by SVM                                 
diff_df.loc[(diff_df.DALX.notna()) & (diff_df.ATLX != diff_df.DALX)] # 1117
```

```python
# Words in S15 vocab but not changed polarity by SVM                                               
diff_df.loc[(diff_df.DALX.notna()) & (diff_df.ATLX == diff_df.DALX) & (diff_df.INVOCAB == True)] # 451 + NaNs
```

```python
# Words in S15 vocab and changed polarity by SVM                                                 
diff_df.loc[(diff_df.DALX.notna()) & (diff_df.ATLX != diff_df.DALX) & (diff_df.INVOCAB == True)] # 160
```

# Aspect specific lexicon - Amazon Electronics

```python
alx = pd.read_csv('data/output/lexicon_table_dalx_asp_10_thres0.7_C10_concat.csv', index_col=['WORD', 'ASP'])
vocab = read_yaml('data/SemEval15_laptop/glove_symdict.yml')
_df = alx.loc[alx.DALX.notna()][['DALX']].unstack()
multi_pol_words = _df.loc[_df.apply(lambda s: len(s.unique()), axis=1) != 1].index.tolist()
multi_pol_words_in_vocab = [w for w in multi_pol_words if w in vocab]
```

```python
# All wrods have different polarity per aspect
alx.loc[multi_pol_words]
```

```python
# Wrods have different polarity per aspect and in corpus vocabulary
alx.loc[multi_pol_words_in_vocab]
```
